```python
#Revisiting Some familiar graph
#Example_1 ------------------------------------------------------------------------>>>>>>>>>>>
#Count plot-------------------->>>>>>

#How many cars of each bosy types are present in the car prices dataset?
# plt.figure(figsize = (6 , 4))
sns.countplot(data = df , x= 'carbody' )
# plt.title('Number of cars by car body type')
# plt.xlabel('Car body type')
# plt.ylabel('count')
```

```python
plt.figure(figsize = (6 , 4))
sns.countplot(data = df , y = 'carbody' , order = df['carbody'].value_counts().index ) #descending
plt.title('Number of cars by car body type')
plt.ylabel('Car body type')
plt.xlabel('count')
```

```python
plt.figure(figsize = (6 , 4))
sns.countplot(data = df , y = 'carbody' , order = ['wagon' , 'hardtop' , 'hatchback' , 'sedan' , 'convertible']  ) #descending
plt.title('Number of cars by car body type')
plt.ylabel('Car body type')
plt.xlabel('count')
```

```python
#Example_3---------------------------------------->>>>>>>>>>>>.
# Bar Graph
# what is average engine size for each car body type in the car prices dataset?
plt.figure(figsize = (6 , 4))
sns.barplot(data = df , y = 'carbody' , x = 'enginesize' , estimator = 'mean' ,errorbar = None)
#if u give categorical and numerical value the sns automatically take mean , estimator takes mean.
#if errorbar is noot used a black line appears ,, this black line gives +- error
```

```python
plt.figure(figsize = (6 , 4))
sns.barplot(data = df , y = 'carbody' , x = 'enginesize' , estimator = 'median' ,errorbar = 'sd')
```

```python
#Example_4
# Grouped bar graph
plt.figure(figsize = (6 , 4))
bar_plot = sns.barplot(data = df , y = 'carbody' , x = 'enginesize' , estimator = 'median' ,errorbar = 'sd')
bar_plot.set_xlabel('mean engine size')
bar_plot.set_ylabel('car body type')
bar_plot.set_title('Avg engine size by car body type split over number of doors')
```

```python
# 3example_5 ------------------->>>>>>>
# Grouped bar graph
plt.figure(figsize = (4 ,6))
bar_plt = sns.barplot(data = df , y = 'carbody' , x = 'enginesize' , hue = 'doornumber' , estimator = 'mean' , errorbar = None)
```
