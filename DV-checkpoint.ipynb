```python
#Example_12

# Let's now explore multiple categorical variables using stacked and grouped bar plots

#How are the cars in the data set distributed over doornumber and carbody

```

```python
doornumber_carbody_crosstable = pd.crosstab(df['doornumber'], df['carbody'])
doornumber_carbody_crosstable


#cross table
```

```python
carbody_categories = doornumber_carbody_crosstable.columns
carbody_categories
```

```python
fig , ax = plt.subplots()
bar1 = ax.bar(np.arange(len(carbody_categories)) ,doornumber_carbody_crosstable.loc['four'] )
bar2 = ax.bar(np.arange(len(carbody_categories)) ,doornumber_carbody_crosstable.loc['two'] )
ax.set_xlabel('index of categories')
ax.set_ylabel('count')
ax.set_title('Distribution o number of cars over number of doors and types of car body');
```

```python
fig , ax = plt.subplots()
bar1 = ax.bar(np.arange(len(carbody_categories)) ,doornumber_carbody_crosstable.loc['four'], alpha = 0.5 )
bar2 = ax.bar(np.arange(len(carbody_categories)) ,doornumber_carbody_crosstable.loc['two'] , alpha = 0.5)
ax.set_xlabel('index of categories')
ax.set_ylabel('count')
ax.set_title('Distribution o number of cars over number of doors and types of car body');
```

```python
fig , ax = plt.subplots()
bar1 = ax.bar(np.arange(len(carbody_categories)) - 0.25,doornumber_carbody_crosstable.loc['four'] , width = 0.5 )
bar2 = ax.bar(np.arange(len(carbody_categories)) + 0.25,doornumber_carbody_crosstable.loc['two'] , width =  0.5 )
ax.set_xlabel('index of categories')
ax.set_ylabel('count')
ax.set_title('Distribution o number of cars over number of doors and types of car body');
```

```python
fig , ax = plt.subplots()
bar1 = ax.bar(np.arange(len(carbody_categories)) - 0.2,doornumber_carbody_crosstable.loc['four'] , width = 0.4)
bar2 = ax.bar(np.arange(len(carbody_categories)) + 0.2,doornumber_carbody_crosstable.loc['two'] , width =  0.4)
ax.set_xlabel('index of categories')
ax.set_ylabel('count')
ax.set_title('Distribution o number of cars over number of doors and types of car body');
```

```python
fig , ax = plt.subplots()
bar1 = ax.bar(np.arange(len(carbody_categories)) - 0.2,doornumber_carbody_crosstable.loc['four'] , width = 0.4 , label = 'four')
bar2 = ax.bar(np.arange(len(carbody_categories)) + 0.2,doornumber_carbody_crosstable.loc['two'] , width =  0.4 , label = 'two')
ax.set_xlabel('index of categories')
ax.set_ylabel('count')
ax.set_title('Distribution o number of cars over number of doors and types of car body')
ax.set_xticks(np.arange(len(carbody_categories)))
ax.set_xticklabels(carbody_categories);
ax.legend();
```

```python
# example 14
# Lets now try to draw different histograms for two door and four door cars

two_door_df = df[df['doornumber'] == 'two']
four_door_df = df[df['doornumber'] == 'four']
```

```python
plt.hist(two_door_df['price'] , label = 'two')
plt.hist(four_door_df['price'] , label = 'four')
plt.xlabel('price')
plt.ylabel('number of cars')
plt.title('Distribution of prices of cars based on number of doors')
plt.legen
```

```python
#example_15 --------------------->>>>

#An alternate way of representing the same information in a stacked histogram
plt.hist([two_door_df['price'], four_door_df['price']] , alpha = 0.5 , label = ['two door' , 'four door'], stacked = True)
plt.xlabel('price') #stacked = True (it will put two graphs one on the toop of the other)
plt.ylabel('number of cars')
plt.title('Distribution of prices of cars based on number of doors')
plt.legend();
```

```python
fig , axs = plt.subplots(1 , 2 , figsize = (8 , 4))
#1 , 2 = 1 row and 2 column means 2 subplots placed in a row and there are 2 columns
```

```python
fig , axs = plt.subplots(1 , 2 , figsize = (8 , 4))
# 8 = width of the figure in inches 
# 4 = height of the figure in inches
axs[0].hist(two_door_df['price'], bins = 20 , color = 'blue')
axs[0].set_title('Histogram of prices of two door cars')

# axs[0] = mean the subplot on the left side

axs[0].set_xlabel('price')
axs[0].set_ylabel('frequency')
axs[0].tick_params(axis = 'x' , rotation = 45)



axs[1].hist(four_door_df['price'], bins = 20 , color = 'green')
axs[1].set_title('Histogram of prices of four door cars')

# axs[1] = mean the subplot on the right side

axs[1].set_xlabel('price')
axs[1].set_ylabel('frequency')
axs[1].tick_params(axis = 'x' , rotation = 45)

plt.suptitle('Histogram of prices of two door and four door cars')  #suoer title

plt.tight_layout();
```
